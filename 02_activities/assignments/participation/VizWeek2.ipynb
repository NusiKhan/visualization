```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from PIL import Image
import requests
from io import BytesIO
import seaborn as sns
```

```python
x= np.array(["Luffy", "Zoro", "nami", "usopp"," sanji"])
y1= np.array([110,180,240,99,220])
y2=np.array([170,100,90,120,50])

fig, ax= plt.subplots(figsize=(7,3))

ax.bar(x,y1, color="indigo")
ax.plot(x,y2,color="red")
```

```python
y2_sd=np.std(y2) #std. dev

fig, ax= plt.subplots(figsize=(7,3))


ax.plot(x,y2,color="red")

#add the error info into the axis object

ax.errorbar(x, y2, 
            yerr = y2_sd,
            fmt = "none")

#fmt means format we want just the error bar part and not the rest of it
```

```python


fig, ax= plt.subplots(figsize=(7,3))


ax.plot(x,y2,color="red")


ax.errorbar(x, y2, 
            yerr = y2_sd,
            fmt = "none",
            ecolor= "purple",
            elinewidth=4,
            capsize=6,
            capthick= 4,
            errorevery= 2)

#error every adds error bars ever 2 points
```

```python
fig,ax= plt.subplots(figsize =(7,3))
ax.plot(x,y2, color= "red")
ax_image= fig.add_axes([0.1,0.11,0.15,0.35])

ax_image.imshow(image)
ax_image.axis('off')
```

```python
tips = sns.load_dataset("tips") #seaborn has tips as a sample database
print(tips)
```

```python
sns.set_style('whitegrid')#setting a style and making a line plot with it

sns.lineplot(data= tips,
                x= 'total_bill',
                y= 'tip')

```

```python
# naming the graph as an object
tipgraph = sns.lineplot(data= tips,
                x= 'total_bill',
                y= 'tip')
tipgraph.set(title = 'Tips vs total bill',
             xlabel= 'totalbill $',
             ylabel= 'tip amount$')
```

```python
fig= plt.subplots(figsize= (10,3))

tipgraph= sns.lineplot(data= tips,
                       x= 'total_bill' , y= 'tip', 
                       color= 'hotpink',
                       linestyle= '--')
                    
```

```python
tipgraph = sns.lineplot(data= tips,
                x= 'total_bill',
                y= 'tip', color= 'hotpink',
                linestyle= '--', linewidth = 3, 
                marker = 'o',
                markerfacecolor = 'indigo')
```

```python
tipgraph= sns.scatterplot(data= tips,
                x= 'total_bill',
                y= 'tip', style= 'time', hue= 'day', palette=  ['purple', 'hotpink', 'blue', 'green']) 

tipgraph.set(title= 'tip vs total bill',
             xlabel= 'total bill', ylabel = 'tip amount')

#hue is change color based on day
```

```python
sns.pairplot( data= tips, hue = 'day')
```

```python
daysplot= sns.relplot(data= tips, x='total_bill', y='tip',
                     hue= 'sex', col= 'day',
                     kind='scatter', col_wrap =2)

# columns according to day , hue is colored based on sex varialbele, colwrap puts them undereach other to make it easier to see
```

dynamic data visualisation

```python
import plotly.graph_objects as go


x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph= go.Figure()

graph.add_trace(go.Bar (x=x1, y=y1)
                   )
graph.update_layout(title= 'pirate scores', xaxis_title= 'pirates', yaxis_title= "score")
```

```python
graph =go.Figure()

graph.add_trace(go.Scatter( x=x1, y=y1, 
                           mode= 'markers',
                           marker= dict(
                                size= 15, 
                                color= 'hotpink', 
                                opacity=1, 

                                        line= dict(width=5, color= 'purple'))))

graph.update_layout( title= 'interactive pirate plot', 
                    xaxis_title= 'pirates', 
                    yaxis_title= 'scores',
                    width= 500, height= 500)
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
A= set(['apple', 'banana', 'watermelon'])
B= set(['pumpling', 'blueberry', 'lime', 'apple'])
```

```python
diagram = venn2([A,B], set_labels= ('fruits', 'pies'),
             set_colors= ('blue', 'red'),
             alpha=0.5
             )
diagram.get_label_by_id("10").set_text("n".join(A - B))
diagram.get_label_by_id("11").set_text("n".join(A & B))
diagram.get_label_by_id("01").set_text("n".join(B - A))
```
