```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1= np.arange(50)
y1= np.random.randint(0,75,50)
x2= np.array(["Luffy", "Zoro", "nami", "usopp"," sanji"])
y2= np.array([110,180,240,99,220])
```

# subplots

```python
#now we have 2 plots next to each other

fig, (ax1, ax2) = plt.subplots(ncols= 2, 
                               nrows=1, 
                               figsize= (7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols= 2, 
                               nrows=1, 
                               figsize= (9,3))

font1 = {'family': 'serif', 'size':10}
font2 = {'family': 'serif', 'size':14}

ax1.scatter(x1,y1, marker ='*', color ='red' , label= "pirates")
ax2.bar(x2,y2, label= "one piece", color= 'pink')

ax1.legend( loc= 'lower left')
ax2.legend( loc= 'lower right')

ax1.set_title('pirate attacks', fontdict=font1)
ax2.set_title('pirate ship', fontdict=font1)
```

```python
fig, someaxes= plt.subplot_mosaic([['ax1', 'ax3'],
                                  ['ax2', 'ax3']],
                                  figsize= (7,4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
fig, someaxes= plt.subplot_mosaic([['ax1', 'ax3'],
                                  ['ax2', 'ax3']],
                                  figsize= (7,4), layout= "constrained")

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

#labels
someaxes["ax1"].set_xlabel("A Big Label", fontsize= 18)

someaxes["ax2"].set_xlabel("Another Label", fontsize= 18)

someaxes["ax3"].set_xlabel("label2: 2 fast", fontsize= 18)

#how to fix this mess? use layouts , layout=constraint

```
