```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#create our sample data, making 2 different y variables 

np.random.seed(613)
x= np.arange(50)
y1= np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots( figsize =(5,3)) #figure with axis object inside it

#line plot with y1 and another line plot with y2 inside the axis object
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
fig, ax = plt.subplots( figsize =(5,3)) #figure with axis object inside it

#adding legend 

ax.plot(x,y1, label = "person1")
ax.plot(x,y2, label= "person2")
ax.legend(loc= "lower right",
          frameon = True, 
          fontsize= 12,
          ncol=2, shadow= True)

#ncol =2 puts them side by side in two columns, can split into more if we had more labels
```

```python
fig, ax = plt.subplots( figsize =(5,3)) #figure with axis object inside it


ax.plot(x,y1, label = "person1")
ax.plot(x,y2, label= "person2")

#move legend outside our plot area but still inside the axis

ax.legend( loc= 'upper left', 
          bbox_to_anchor= (1,1))
#bbox 1,1 is the position where on our figure we want to anchor our legend
#at the 1,1 is where we want our legend to be
```

```python
fig, ax = plt.subplots( figsize =(5,3)) 

ax.scatter(x,y1, label = "person1")
ax.scatter(x,y2, label= "person2")
ax.legend(loc = 'lower right')

ax.text(10,95, "this value is important") #adding a text annotation
```

```python
fig, ax = plt.subplots( figsize =(5,3)) 

ax.scatter(x,y1, label = "person1")
ax.scatter(x,y2, label= "person2")
ax.legend(loc = 'lower right')

ax.text(10,95, "this value is important", 
        ha = "center",
        color = "red",
        size = 20) #we can modify our annotation by adding arguments
# ha is horizontal alignment
```

```python
#using transforms 
# 0 to 10 on x axis and then on y axis
#0.5, 0.1 50% of the way across and 10 % up

fig,ax= plt.subplots()

ax.axis ([0,10,0,10])

ax.text (1,5, ". Data (1,5)", transform = ax.transData)
ax.text(0.5, 0.1,". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2,0.2,". Figure: (0.2,0.2)", transform=fig.transFigure)

```

```python
fig, ax = plt.subplots( figsize =(5,3)) 

ax.scatter(x,y1, label = "person1")
ax.scatter(x,y2, label= "person2")
ax.legend(loc = 'lower right')

#we can annotate with arrows

ax.annotate("This is important", xy = (10,95),
            xytext= (20,94),
            arrowprops=dict(facecolor= 'black', shrink=0.005))
```

```python
fig, ax = plt.subplots( figsize =(5,3)) 

ax.scatter(x,y1, label = "person1")
ax.scatter(x,y2, label= "person2")
ax.legend(loc = 'lower right')



ax.annotate("This is important", xy = (10,95),
            xytext= (20,94),
            arrowprops=dict(facecolor= 'pink', arrowstyle= "wedge"))
```

```python
#axis labels x and y axis not the other axis

fig, ax = plt.subplots( figsize =(5,3)) 

ax.scatter(x,y1, label = "person1")
ax.scatter(x,y2, label= "person2")
ax.legend(loc = 'lower right')
#appreance of axis
ax.yaxis.set_major_locator(plt.NullLocator()) #locator to remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) #formatter to remove only labels
```

```python
fig, ax = plt.subplots( figsize =(5,3)) 

ax.scatter(x,y1, label = "person1")
ax.scatter(x,y2, label= "person2")
ax.legend(loc = 'lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #only 3 ticks on the x axis
```

```python
fig, ax = plt.subplots( figsize =(5,3)) 

ax.scatter(x,y1, label = "person1")
ax.scatter(x,y2, label= "person2")
ax.legend(loc = 'lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # interval on the x axis after every 5 unitx
```

```python
fig, ax = plt.subplots( figsize =(5,3)) 

ax.scatter(x,y1, label = "person1")
ax.scatter(x,y2, label= "person2")
ax.legend(loc = 'lower right')

#modify labels on our axis and sepicify the rotation and axis

plt.xticks( rotation=45, 
           ha='right', 
           fontsize= 12,
             font= "serif")
```
