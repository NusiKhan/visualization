```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
#make some sample data

np.random.seed(613)
x= np.arange(50)
y= np.random.randint(0,100,50) # 0 is our min, 100 max, and 50 of those values

#graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
#barchart

fig, ax = plt.subplots( figsize =(5,3))

ax.bar(x,y)
#ax.barh(x,y) to make it horizontal
```

```python
fig, ax = plt.subplots( figsize =(5,3))

ax.plot(x,y)
```

```python
fig, ax = plt.subplots( figsize =(5,3)) 

ax.hist(y) #occurences of each value iin the y cariable
```

```python
#adding labels and titles to a graph
fig, ax = plt.subplots( figsize =(5,3))

ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('years since start')
```

```python
#dictonaries are key value pairs in {}

font1= {'family': 'sans-serif', 'color':'blue', 'size':20}
font2= {'family': 'monospace', 'color': 'green', 'size':14}
```

```python
#using the dictonary to change font in the labels and titles

fig, ax = plt.subplots( figsize =(5,3))

ax.plot(x,y)

ax.set_title('Total growth over time', fontdict= font1)
ax.set_ylabel('Total growth' , fontdict = font2)

ax.set_xlabel('years since start', fontdict = font2)
```

```python
#Experimenting with placements for labels
fig, ax = plt.subplots( figsize =(5,3))

ax.plot(x,y)

ax.set_title('Total growth over time', fontdict= font1, loc ='left')
ax.set_ylabel('Total growth' , fontdict = font2, loc= 'bottom')

ax.set_xlabel('years since start', fontdict = font2, loc = 'right')
```

```python
#modified scatterplot
fig, ax = plt.subplots( figsize =(5,3))
ax.scatter(x,y,
           marker ='*',
           color= 'indigo')

#markers can be different shapes *, >, o, + etc
```

```python
#creating a line graph with changes in the shape of marker, color, width and style 
fig, ax = plt.subplots( figsize =(5,3))
ax.plot(x,y,
           marker ='+',
           color= 'indigo',
           linestyle = '--',
           linewidth = 2)

```

```python
fig, ax = plt.subplots( figsize =(5,3))
ax.plot(x,y,
           marker ='*',
           color= 'indigo',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')

ax.grid(axis = 'y') # () blank will give both gridlines
```

```python
fig, ax = plt.subplots( figsize =(5,3))
ax.scatter(x,y,
           marker ='*',
           color= 'indigo',
           linestyle = '--',
           linewidth = 2,
           )

ax.grid(axis = 'y') # () blank will give both gridlines
```

```python
#trying to create a box plot with the things we learnt above
fig, ax = plt.subplots(figsize=(5, 3))


ax.boxplot(y)


ax.set_title('Boxplot ')
ax.set_xlabel('x', fontdict=font2)
ax.set_ylabel('y', fontdict= font1)
# Show the plot
plt.show()
```
